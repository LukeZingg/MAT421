**17.1 - Interpolation Problem Statement**



*  Interpolation is the process of assigning a function to a set of data that passes through each point in the data via an *interpolation function*.
*   Interpolation, unlike regression, does not require a pre-existing model for the data. That said, the nature of our data may influence how we define our interpolation function, such as suggesting our function should be "smooth" (i.e. the derivative of the interpolation function is continuous).

**17.2 - Linear Interpolation**



*   Linear interpolation begins by ordering our data from least to greatest x-values.
*   It then assumes that an estimated point lies on the line joining the nearest data points to its left and right.
*   Thus, the linear interpolation for some point x where x_i < x < x_i+1 is f(x) = y_i + [(y_i+1 - y_i)(x - x_i)] / (x_i+1 - x_i)



```python
#Linear Interpolation:

from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [2, 1, 5, 4]

f = interp1d(x, y)
y_hat = f(1.5)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.3 - Cubic Spline Interpolation**



*   Cubic Spline Interpolation generates a function comprised of pieceswise cubic functions.
*   The interpolation will have n-1 cubic functions for n-many datapoints, with each function having 4 coefficients.
*   This type of interpolation proves useful when we are modeling data where we have reason to believe the derivative of our interpolation should be continuous, such as a vehicle's velocity.


```python
#A demonstration of cubic spline interpolation:

from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import numpy as np

x = [0, 2, 4, 6, 8, 10, 12]
y = [2, 3, 1, 7, 1, 2, 1]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 12, 100)
y_new = f(x_new)

plt.figure(figsize = (8,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
